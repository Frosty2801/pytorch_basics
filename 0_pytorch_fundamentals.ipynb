### Reproducibility


```python
import torch

random_tensor_a = torch.rand(3, 4)
random_tensor_b = torch.rand(3, 4)

print(random_tensor_a)
print(random_tensor_b)
print(random_tensor_a == random_tensor_b)
```

```python
# random but reproducible tensor
import torch

# set random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_c = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tenser_d = torch.rand(3, 4)

print(random_tensor_c)
print(random_tenser_d)
print(random_tensor_c == random_tenser_d)
```

### Putting tensors (and models) on the gpu

For more faster computations


```python
import torch
tensor = torch.tensor([2, 4, 6])

# tensor not on gpu
print(tensor, tensor.device)
```

```python
# move to gpu
gpu_tensor = tensor.to(device)
gpu_tensor
```

```python
cpu_tensor = gpu_tensor.cpu().numpy()
cpu_tensor
```
